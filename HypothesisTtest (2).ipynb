```python
#LifestyleScore
# importing the pandas library 
import pandas as pd 
csv_file = "C:\\Users\\gisfe\\Downloads\\QUESTIONNAIRE DATASET.csv"
# reading the csv file 
df = pd.read_csv(csv_file) 



# Define a mapping for the values
value_mapping1 = {'Less than 100': 1, '100 - 250': 2, '250 - 500': 3, 'More than 500': 4}


# Map the values in the 'Column' based on the defined mapping
df['How much do you spend on food every day on an average?'] = df['How much do you spend on food every day on an average?'].map(value_mapping1)
#df['Background'] = df['How much do you spend on food every day on an average?'].map(value_mapping1)
# Print the modified DataFrame
print(df)

# writing into the file 
df.to_csv("C:\\Users\\gisfe\\Downloads\\QUESTIONNAIRE DATASET.csv", index=False) 
  
# Map the values in the 'Column' based on the defined mapping
df['How much do you spend on travel every week on an average?'] = df['How much do you spend on travel every week on an average?'].map(value_mapping1)


# Define a mapping for the values
value_mapping2 = {'Less than 1000': 1, '1000 - 2500': 2, '2500 - 5000': 3, 'More than 5000': 4}

# Map the values in the 'Column' based on the defined mapping
df['How much do you spend on clothes and accessories every month on an average?'] = df['How much do you spend on clothes and accessories every month on an average?'].map(value_mapping2)
# Map the values in the 'Column' based on the defined mapping
df['How much do you spend on daily neccesseties every month on an average?'] = df['How much do you spend on clothes and accessories every month on an average?'].map(value_mapping2)



  
# updating the column value/data 
    
df['LifestyleScore'] = 1 + 2 * (df['Which of the following statements best describes how you purchase items'] == 'I consider products from different companies before buying, I consider only affordable options')
+ 3 * (df['Which of the following statements best describes how you purchase items'] == "I don't consider buying unnecessary products")
+ 1 * (df['Which of the following statements best describes how you purchase items'] == "I am loyal to certain brands and products, I don't consider other affordable options")
+ 1 * (df['How much do you spend on food every day on an average?'])+ 1 * (df['How much do you spend on travel every week on an average?'])
+ 1 * (df['How much do you spend on clothes and accessories every month on an average?'])+ 1* (df['How much do you spend on daily neccesseties every month on an average?'])  


# writing into the file 
df.to_csv("C:\\Users\\gisfe\\Downloads\\QUESTIONNAIRE DATASET.csv", index=False) 
  
print(df)
```

```python
df_score = df[ 
    (df['FinancialLiteracyScore'].notnull()) & 
    (df['LifestyleScore'].notnull())
][['FinancialLiteracyScore','LifestyleScore']]
df_score
```

```python
from scipy.stats import ttest_rel

test_statistic, p_value = ttest_rel(df_score['FinancialLiteracyScore'], df_score['LifestyleScore'],alternative='less')

print("test statistic= ",test_statistic)
print("P value= ",p_value)

if p_value < 0.05:
    print("\nWe Reject Null Hypothesis")
else:
    print("\nWe Accept Null Hypothesis")
```

```python
df_score = df[ 
    (df['FinancialLiteracyScore'].notnull()) & 
    (df['EducationScore'].notnull())
][['FinancialLiteracyScore','EducationScore']]
df_score
```
